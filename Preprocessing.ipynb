# Tahap Prepocessing

## Punctuation Removal

```python
print(df.dtypes)
```

```python
import string
string.punctuation

# defining the function to remove punctuation
def remove_punctuation(text):
    punctuationfree = re.sub(r'[^a-zA-Z\s]', '', text)
    return punctuationfree
    
# storing the puntuation free text
df['Berita']= df['Berita'].apply(lambda x:remove_punctuation(x))
df.head()
```

```python
# Save to CSV
df.to_csv('punctuation_removal_all.csv', index=False)
```

## Case Folding

```python
df = df.map(lambda x: x.lower() if isinstance(x, str) else x)
df.head(10)
```

```python
# Save to CSV
df.to_csv('case_folding_all.csv', index=False)
```

## Tokenization

```python
import nltk
from nltk.tokenize import word_tokenize

df['Berita'] = df['Berita'].apply(word_tokenize)
df.head()
```

```python
# Save to CSV
df.to_csv('tokenization_all.csv', index=False)
```

## Stopword Removal

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('indonesian'))
df['Berita'] = df['Berita'].apply(lambda x: [word for word in x if word not in stop_words])
df.head()
```

```python
# Save to CSV
df.to_csv('stopword_removal_all.csv', index=False)
```

## Stemming

```python
# Initialize the Stemmer Factory
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# List of words to exclude from stemming
exception_words = {"pendidik", "pengasuh", "pembantu"}

# Define a function to perform customized stemming
def stem_words(words):
    stemmed_words = []
    for word in words:
        if word in exception_words:  # Check if the word is in the exception list
            stemmed_words.append(word)  # Keep the word as is
        else:
            stemmed_words.append(stemmer.stem(word))  # Stem other words
    return stemmed_words

# Apply the function to the 'text' column and create a new column 'stemmed_text'
df['Berita'] = df['Berita'].apply(stem_words)

df.head(10)
```

```python
# Save to CSV
df.to_csv('Preprocessing/Result Preprocessing.csv', index=False)
```
