### Read File

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
filename = "Word2Vec/word_frequencies_each_category.csv"
```

```python
df = pd.read_csv(filename, header=0)
df.head(5)
```

### WordCloud Each Category

```python
# Exclude unwanted words
excluded_words = {
    "anak", "korban", "laku", "duga", "polisi"}

# Create word clouds for each category
categories = df["Category"].unique()

for i, category in enumerate(categories, 1):
    category_data = df[df["Category"] == category]
    words_freq = {row["Word"]: row["Frequency"] for _, row in category_data.iterrows() if row["Word"] not in excluded_words}
    
    # Generate word cloud
    wordcloud = WordCloud(width=500, height=500, background_color="white").generate_from_frequencies(words_freq)
    
    # Plot
    print(category)
    plt.figure(figsize=(15, 10))
    plt.subplot(2, 3, i)  # Adjust layout if needed
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.tight_layout()
    plt.show()

```
