```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Word2Vec/Loss.csv')
data.head()
```

```python
# Buat plot grafik
# plt.figure(figsize=(10, 6))
plt.plot(data['Epoch'], data['Loss'], linestyle='-')
plt.title('Loss Training Word2Vec')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.tight_layout()

# Tampilkan grafik
plt.show()
```
