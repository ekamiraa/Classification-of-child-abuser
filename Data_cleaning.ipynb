```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
filename = "Dataset labelling/Dataset Berita Kekerasan Anak.csv"
```

```python
df = pd.read_csv(filename, header=0)
df.head()
```

```python
df.shape
```

```python
df['Label'] = df['Label'].str.lower()
df.describe()
```

### Manage Duplicated Data

```python
df.isnull().sum()
```

```python
# Check for duplicates
duplicates_counts = df.duplicated(subset='Judul').sum()
print('Jumlah data duplikasi = ', duplicates_counts)

# Show Duplicated data
df[df.duplicated(subset='Judul')]
```

```python
# Delete duplicates
df.drop_duplicates(subset='Judul', inplace=True)
df.reset_index(drop=True, inplace=True)

df.shape
```

### Delete Columns That Are Not Featured

```python
df.drop(['Judul', 'Tanggal', 'Link'], axis=1, inplace=True)
```

```python
df.head()
```

### Label Encoded

```python
print(df['Label'].unique())
```

```python
#  Add new column and set value to category of label column
#  keluarga = 0, pengasuh = 1, tenaga pendidik = 2, teman = 3, orang asing = 4 
label_encode = {'keluarga': 0, 'pengasuh': 1, 'tenaga pendidik': 2, 'teman': 3, 'orang asing': 4}
df['Category_id'] = df['Label'].map(label_encode)
df
```

```python
# Save to CSV
df.to_csv('Dataset Cleaning/Result Dataset Cleaning.csv', index=False)
```

### Exploratory Data Analysis

```python
label_counts = df.groupby('Label').size()
print(label_counts)
```

```python
# Create Visualization
plt.figure()
bars = label_counts.plot(kind='bar', color='skyblue')
plt.title('Distribusi Pelaku Kekerasan Anak')
plt.ylabel('Jumlah Kasus', labelpad=20.0)
plt.xlabel('Kategori Pelaku', labelpad=20.0)
plt.xticks(rotation=360)

# Add text above graphics
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
             f'{int(bar.get_height())}', 
             ha='center', va='bottom', color='black')

plt.show()
```
